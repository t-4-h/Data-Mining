```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
```

```python
URL = 'https://www.politifact.com/factchecks/list/?ruling=mostly-true'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
results = soup.find(id='o-listicle__list')
#print(results.prettify())
```

```python
items_all = soup.find_all('li', class_='o-listicle__item')
for item in items_all:
    print(item, end='\n'*2)
```

```python
true_level = 'mostly-true'
```

```python
result_text = []
for item in items_all:
    author_item = item.find('a', class_='m-statement__name')
    whenwhere_item = item.find('div', class_='m-statement__desc')
    content_item = item.find('div', class_='m-statement__quote')
    if None in (author_item, whenwhere_item, content_item):#, company_item, location_item
        continue
    print(author_item.text.strip(), '\t', whenwhere_item.text.strip(), '\t', content_item.text.strip(), '\t', true_level)
    text = author_item.text.strip(), '\t', whenwhere_item.text.strip(), '\t', content_item.text.strip(), '\t', true_level
    result_text.append(text)
    #print(whenwhere_item.text.strip())
    #print(content_item.text.strip())
    #print()
```
